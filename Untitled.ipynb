```python
# 1. Define custom functions for Q1, Q3, and IQR
def q1(x): return x.quantile(0.25)
def q3(x): return x.quantile(0.75)
def iqr(x): return x.quantile(0.75) - x.quantile(0.25)

# 2. Group by 'Source' and aggregate the 'Length' column
summary_df = df.groupby('Source')['Length'].agg(
    median='median',
    iqr=iqr,
    min='min',
    max='max',
    q1=q1,
    q3=q3
).reset_index().sort_values(by='median', ascending = True)

summary_df

```
